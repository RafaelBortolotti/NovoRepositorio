## Agrupamento de Valores:
### 1 - Quais são as Top 5 Plataformas mais comercializadas

```python
# Selecionando as variáveis que serão análisadas
Platform = df.filter(['Platform', 'Global_Sales'])

# Criando um agrupamento por Plataforma
Platform = Platform.groupby(['Platform']).sum()

# Ordenando os dados por Global_Sales
Platform = Platform.sort_values('Global_Sales', ascending=False).head(5)

# Adicionando um gráfico de barras para visualização
Platform.plot(kind='bar')
plt.suptitle('Top 5 Plataformas em Vendas')
plt.xlabel('Plataformas')
plt.ylabel('Vendas Global')
plt.show()

# incluir os rótulos de dados

```
